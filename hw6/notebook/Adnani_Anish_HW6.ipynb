<b> 1.(e) XGBoost and Model Trees In the case of a univariate tree, only one input dimension is used at a tree split. In a multivariate tree, or model tree, at a decision node all input dimensions can be used and thus it is more general. In univariate classification trees, majority polling is used at each node to determine the split of that node as the decision rule. In model trees, a (linear) model that relies on all of the variables is used to determine the split of that node (i.e. instead of using X j > s as the decision rule, one has P j β j X j > s . as the decision rule). Alternatively, in a regression tree, instead of using average in the region associated with each node, a linear regression model is used to determine the value associated with that node. One of the methods that can be used at each node is Logistic Regression. Because the number of variables is large in this problem, one can use L 1 -penalized logistic regression at each node. You can use XGBoost to fit the model tree. Determine α (the regularization term) using cross-validation. Train the model for the APS data set without compensation for class imbalance. Use one of 5 fold, 10 fold, and leave-one-out cross validation methods to estimate the error of your trained model and compare it with the test error.
```python
xgb_errors = {}
all_alphas = 10**np.linspace(10,-2,20)*0.5
xgb_X_columns = list(train_X.columns)
xgb_Y_columns = ['class']

for alpha in all_alphas:
    folds = StratifiedKFold(n_splits=10, shuffle=False)
    cross_val_error = []
    
    
    for train_indx, val_indx in folds.split(new_train_X, new_train_Y):
        xgbTrainingX = []
        xgbTrainingY = []
        
        for it in train_indx:
            xgbTrainingX.append(new_train_X.iloc[it])
            xgbTrainingY.append(new_train_Y.iloc[it])
            
        xgbTrainingX = pd.DataFrame(data=xgbTrainingX, columns=xgb_X_columns)
        xgbTrainingY = pd.DataFrame(data=xgbTrainingY, columns=xgb_Y_columns)
        
        xgbValidationX = []
        xgbValidationY = []
        
        for it in val_indx:
            xgbValidationX.append(new_train_X.iloc[it])
            xgbValidationY.append(new_train_Y.iloc[it])
        xgbValidationX = pd.DataFrame(data=xgbValidationX, columns=xgb_X_columns)
        xgbValidationY = pd.DataFrame(data=xgbValidationY, columns=xgb_Y_columns)
        
        
        xgb_model_ = XGBClassifier(eval_metric='auc', reg_alpha=alpha)
        xgb_model_ = xgb_model_.fit(xgbTrainingX, xgbTrainingY)
        
        
        validation_Y_pred = xgb_model_.predict(xgbValidationX)
        error = 1 - accuracy_score(xgbValidationY, validation_Y_pred)
        cross_val_error.append(error)
    xgb_errors[alpha] = np.mean(cross_val_error)

cross_val_rows = [[x, xgb_errors[x]] for x in xgb_errors]
cross_val_xgb_df = pd.DataFrame(data=cross_val_rows, columns=['alpha', 'Cross Validation Error'])
```
